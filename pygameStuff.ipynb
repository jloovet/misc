```python
import pygame
```

```python
pygame.init()
```

```python
screen_size = (500,500)
```

```python
screen = pygame.display.set_mode(screen_size)
```

```python
#need event loop
while True :
    for event in pygame.event.get() :
        print(event)
        
    pygame.draw.rect(screen, (255,0,0), pygame.Rect(50,60, 200, 300))
    #flip to update screen
    pygame.display.flip()
```
